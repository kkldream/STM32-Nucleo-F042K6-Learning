```python
n = 1
t = 0
print(f'\n{n:2}:  ', end='')
n += 1
for i in range(16):
  for b in range(11):
    print(f'{i:2}[{b:2}]  ', end='')
    t += 1
    if t == 8:
      t = 0
      print(f'\n{n:2}:  ', end='')
      n += 1
```
